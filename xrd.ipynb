```python
from pymatgen import Structure
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer, PointGroupAnalyzer
from jupyter_jsmol.pymatgen import quick_view
from pymatgen.ext.matproj import MPRester
import math
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
with MPRester("iLIbKv6XbyJaj3FOZJZ6") as m:
    # Structure for material id
    struct = m.get_structure_by_material_id("mp-165")
    
    # Visualization
    view = quick_view(struct)
    display(view)
```

```python
print("      Diffractogramme correspondant au Si pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å")
result = XRDCalculator(wavelength=1.5406, symprec=0, debye_waller_factors=None)
result.show_plot(struct, two_theta_range=(0, 39))
tatata = result.get_pattern(struct, scaled=True, two_theta_range=(0, 39))
print(tatata.hkls)
```
